## Task3: Display an Image with Different Interpolation Methods in Matplotlib

- Import matplotlib.pyplot and matplotlib.image.
- Read an image from the example path 'example/path/to/image.jpg'.
- Display the image using the following interpolation methods: 'none', 'nearest','bilinear', 'bicubic', 'spline16', and 'lanczos'.
- Create subplots to show the image with each interpolation method, labeling each subplot with the method name.

```python
import matplotlib.pyplot as plt
import matplotlib.image as myimg

image_path1="./IMAGE_DATA/Dianosars/420.jpg"
original_image=myimg.imread(image_path1)

interpolation_methods = ['none', 'nearest', 'bilinear', 'bicubic', 'spline16', 'lanczos']

fig, axes = plt.subplots(2, 3, figsize=(15, 10)) #subplots

for i, method in enumerate(interpolation_methods): #interpolation ko use
    ax = axes[i // 3, i % 3]
    ax.imshow(original_image, interpolation=method)
    ax.set_title(method)
    ax.axis('off')

plt.show()
```

#### Resize and Create a Thumbnail of an Image with Pillow

- Load the image from the example path.
- Print the original size of the image.
- Create a thumbnail of the image with a maximum size of 100x100 pixels.
- Print the size of the thumbnail.
- Resize the image to 200x200 pixels and print the new size.

```python
from PIL import Image
import matplotlib.pyplot as plt

image_path = './IMAGE_DATA/Flowers/622.jpg'
image = Image.open(image_path)

original_size = image.size
print("Original size:", original_size)

thumbnail_size = (100, 100)
image.thumbnail(thumbnail_size)

thumbnail_size = image.size
print("Thumbnail size:", thumbnail_size)

resized_image = image.resize((200, 200))

resized_size = resized_image.size
print("Resized image size:", resized_size)
```

```python
#thumbnail
plt.subplot(1,3,1)
plt.imshow(image)
plt.axis("off")

plt.subplot(1,3,2)
plt.imshow(resized_image) #resized image is the thumbnail image
plt.axis("off")
```

#### Flip an Image Horizontally and Vertically with Pillow

- Load the image from the example path.
- Create horizontal and vertical flipped versions of the image.
- Display the original, horizontal flip, and vertical flip using Matplotlib.

```python
from PIL import Image
import matplotlib.pyplot as plt

image_path = './IMAGE_DATA/Elephants/520.jpg'
image = Image.open(image_path)

horizontal_flip = image.transpose(Image.FLIP_LEFT_RIGHT)
vertical_flip = image.transpose(Image.FLIP_TOP_BOTTOM)

plt.figure(figsize=(10, 5))

plt.subplot(1, 3, 1)
plt.imshow(image)
plt.title("Original")
plt.axis("off")

plt.subplot(1, 3, 2)
plt.imshow(horizontal_flip)
plt.title("Horizontal Flip")
plt.axis("off")

plt.subplot(1, 3, 3)
plt.imshow(vertical_flip)
plt.title("Vertical Flip")
plt.axis("off")

plt.show()
```
